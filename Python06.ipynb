#Plotando gr√°ficos a partir de f√≥rmulas
At√© agora, s√≥ plotamos gr√°ficos usando pontos que n√≥s mesmo fornecemos. Os pontos, tanto do eixo x quanto do eixo y foram fornecidos, como por exemplo os meses e as temperaturas m√©dias. Vamos ver como o Python pode nos ajudar a plotar gr√°ficos atrav√©s de f√≥rmulas matem√°ticas.

##Lei da Gravita√ß√£o Universal
De acordo com a lei de Newton, dois corpos que possuem massa, sofrem a a√ß√£o de uma for√ßa de atra√ß√£o proporcional ao produto de suas massas e inversamente proporcional a sua dist√¢ncia.

$F = \frac{Gm_1 * Gm_2}{r^2}$,

onde r √© a dist√¢ncia entre os dois corpos e G √© a constante gravitacional. N√≥s queremos ver o que acontece com a for√ßa quando a dist√¢ncia entre os corpos dimunui.

A massa do corpo 1 √© 0,5kg e a do corpo 2 √© 1,5kg. O valor da constante gravitacional √© 6,674 x 10 N m kg. Agora, vamos calcular a for√ßa entre os dois corpos em 19 dist√¢ncias diferentes: 100m, 150m, 200m, 250m, 300m e assim vai at√© 1000m. 

```
from pylab import plot, legend, title, xlabel, ylabel

def draw_graph(x, y):
  plot(x, y, marker='o')
  xlabel("Dist√¢ncia em metros")
  ylabel("For√ßa gravitacional em newtons")
  title("For√ßa gravitacional e dist√¢ncia")
  show()

def generate_F_r():
  # Gerando os valores para r
  r = range(100, 1001, 50)
  # Lista vazia para armazenar os valores calculados de F
  F = []

  # Constant, G
  G = 6.674*(10**-11) 
  # duas massas
  m1 = 0.5
  m2 = 1.5

  # Calcula a for√ßa e adiciona na lista F
  for dist in r:
    force = G*(m1*m2)/(dist**2)
    F.append(force)
  
  # Chamada da fun√ß√£o que desenha o gr√°fico
  draw_graph(r, F)

```

```
generate_F_r()
```

Esse foi o c√≥digo com mais linhas que vimos at√© agora, mas n√£o deve ser dif√≠cil de entender, uma vez que vimos cada coisa que aparece nesse programa. Mas, vamos ver com mais calma.

Foi definido duas fun√ß√µes, uma para desenhar o gr√°fico e outra para calcular a for√ßa e armazenar o resultado em uma lista.

Na fun√ß√£o draw_graph(x,y), todas as linhas que est√£o dentro da fun√ß√£o, n√≥s acabamos de ver. Como por exemplo colocar t√≠tulo, r√≥tulos e usar a fun√ß√£o plot(), que recebe os valores de x e y, mais o caractere de marca√ß√£o do ponto no gr√°fico.
No nosso gr√°fico, x √© a for√ßa e y a dist√¢ncia entre os corpos.

Quando criamos a fun√ß√£o generate_F_r(), fica claro que estamos criando uma fun√ß√£o que ir√° gerar o valor de F, que √© a for√ßa e o valor de r, que √© a dist√¢ncia entre os dois corpos.

Sabemos que teremos 19 dist√¢ncias, que inicia com 100 e termina em 1000, variando de 50 em 50 metros.
Por isso, usamos um range onde fornecemos o valor inicial que √© 100, o valor final que √© 1000 e qual o intervalo ele usar√° para calcular o pr√≥ximo valor, que √©  50.

Depois, criamos a lista vazia F, que vamos reservar para armazenar o c√°lculo da for√ßa, que faremos em seguida.

Com os valores atribu√≠dos √†s vari√°veis G, m1 e m2, j√° podemos calcular a for√ßa F.

Como temos que calcular a for√ßa para cada dist√¢ncia em r, usamos um for que vai percorrer a lista r e calcular a for√ßa para cada item da lista e armazenar o resultado na lista F.

Depois de preenchermos F, j√° temos as duas listas que precisamos para plotar o gr√°fico. Ent√£o chamamos a fun√ß√£o que criamos, draw_graph(x,y), onde x receber√° a lista r que cont√©m as dist√¢ncias entre os corpos e y receber√° a lista F, que cont√©m a for√ßa calculada para cada dist√¢ncia, considerando as massas dos dois corpos.

Use os valores -1, 1, 2, 3, 4 e 5 para x.

```
#O gr√°fico de x^2 + 2x + 1 = 0 cont√©m um ponto - (-1, 0); farei a visualiza√ß√£o da fun√ß√£o f(x) = x^2 + 2x + 1.

def f(x):
  return x**2 + 2*x + 1

def draw_graph(x, y):
  plot(x, y, marker = "o")
  xlabel("Eixo x")
  ylabel("Eixo y")
  title("Gr√°fico de y = x¬≤ + 2x + 1")
  show()

x = [-1, 1, 2, 3, 4, 5]
y = []
for i in x:
  y.append(f(i))

draw_graph(x,y)
```

```
import matplotlib.pyplot as plt

def grafico_barras(dados, rotulos):
  # N√∫mero de barras
  num_bars = len(dados)
  #posi√ß√µes das barras no eixo x
  posicoes = range(1, num_bars+1)
  plt.barh(posicoes, dados, align='center')
  # Positiona os r√≥tulos para cada barra
  plt.yticks(posicoes, rotulos)
  plt.xlabel("Total")
  plt.ylabel("Categoria")
  plt.title("Gastos mensal")
  plt.show()


categorias = ["alimentacao", "transporte", "lazer", "telefone"]
total = [300, 430, 250, 120]

grafico_barras(total, categorias)

```
